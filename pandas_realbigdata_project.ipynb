### Amazon Global Sales Analysis (big data project)

```python
import pandas as pd
```

```python
df= pd.read_csv( "cleaned_dataset.csv" )
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
total_revenue=df['total_revenue'].sum()
total_profit=df['profit'].sum()
print(f"Total Revenue: {total_revenue} crore")
print(f"Total Profit: {total_profit} crore")
```

```python
top_categories = df.groupby('product_category')['quantity_sold'].sum().sort_values(ascending=False)
print(top_categories)
```

```python
profitable_categories = df.groupby('product_category')['profit'].sum().sort_values(ascending=False)
print(profitable_categories)
```

```python
region_sales = df.groupby('customer_region')['total_revenue'].sum().sort_values(ascending=False)
print(region_sales)
```

```python
rating_analysis = df.groupby('product_category')['rating'].mean().sort_values(ascending=False)
rating_analysis
```

```python
payment_method = df['payment_method'].value_counts()
payment_method
```

```python
df.groupby('discount_percent')['quantity_sold'].mean().sort_values(ascending=False)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
sales_trend = df.groupby(df['order_date'].dt.month)['total_revenue'].sum()
sales_trend
```

```python
top_products = df.groupby('product_id')['quantity_sold'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
df['profit_margin'] = df['profit'] / df['total_revenue']* 100
df[['profit_margin']].describe()
```

```python
pivot1 =pd.pivot_table(df, values='total_revenue', index='customer_region', columns='product_category', aggfunc='sum', fill_value=0)
pivot1
```

```python
pivot2 = pd.pivot_table(df, values='rating', index='product_category', columns='customer_region', aggfunc='sum', fill_value=0)
pivot2
```

```python
df.groupby('product_category').agg({'quantity_sold':'sum', 'total_revenue':'sum', 'profit':'sum'}).sort_values(by='total_revenue', ascending=False)
```

```python
df.groupby('product_category').agg({'total_revenue':['sum','mean' , 'max'] ,'profit':['sum', 'mean'], 'rating':['mean','sum','max']}).sort_values
```

```python
pd.crosstab(df['customer_region'], df['payment_method'])
```
